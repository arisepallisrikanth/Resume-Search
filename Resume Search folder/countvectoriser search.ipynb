```python
#creating an object which returns ngrams of words 1 to 9
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(ngram_range=(1,10))
```

```python
# importing dataset
import pandas as pd
df=pd.read_csv(r'Downloads/UpdatedResumeDataSet.csv')

c=list(df["Resume"])
len(c)
```

```python
#removing duplicates
df1=df.drop_duplicates(inplace=False)
C=list(df1["Resume"])
len(C)
```

```python
vectorizer = CountVectorizer(ngram_range=(1,10))
```

```python
#fitting our resume documents
vectorizer.fit(C)
```

```python
#indices assigned to each one of ngrams from 1 to 10 
vectorizer.vocabulary_
```

```python
#looking at the vector representation of each document
vector = vectorizer.transform(C)
vector
print(vector[1])
```
